# Memperbaiki dataset yang NaN

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = 'vgsales.csv'
df = pd.read_csv(data)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Year'].fillna(value=df['Year'].mean())
```

```python
df.head(10)
```

```python
df['Platform'].unique().size
```

```python
df['Genre'].unique().size
```

```python
le = LabelEncoder()
df = df[['Rank', 'Name', 'Platform', 'Year', 'Genre']]
df['Platform'] = le.fit_transform(df['Platform'])
df['Genre'] = le.fit_transform(df['Genre'])
df.head()
```

```python
df.to_csv('vgsales-fix.csv', index=False)
df.head()
```

# Random Splitting

```python
df = pd.read_csv('vgsales-fix.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0)
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data val: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')
```
